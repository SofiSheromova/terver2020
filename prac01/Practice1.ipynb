### Основные формулы
$$P = \frac{m}{n} = \frac{\textrm{число благоприятных исходов}}{\textrm{число всех исходов}}$$


Число способов выбрать $k$ объектов из $n$ различных объектов.
$$C_{n}^{k} = \frac{n!}{k!(n-k)!}$$

Число перестановок
$$P_n = n!$$

### Задачи на пару

**Задача 1.** 

У нас есть правильный кубик. Какова вероятность выпадения:
* При двух бросках, оба раза выпала единица
* Сумма 2х бросков - 7
* Сумма 3х бросков - 16


**Задача 2 (1 балл).** 

В программе к экзамену по теории вероятностей 75 вопросов. Студент знает 50 из них. В билете 3 вопроса. Найдите вероятность того, что студент знает хотя бы два вопроса из вытянутого им билета.

**Задача 3. (TED)** 

За закрытой дверью сидят два ребенка. Они могут быть равновероятно девочкой и мальчиком. Какова вероятность, что там 2 мальчика, если, 
* там точно есть один мальчик
* мы слышали крик за дверью, и этот крик точно принадлежал мальчику.

P.S. Любой ребенок может кричать или не кричать с равной вероятностью.

**Задача 4.** 

У нас есть 3 карточки: одна чисто красная, вторая чисто синяя, третья - одна грань красная, другая - синяя. Мы взяли рандомную карту. Она оказалась синяя. Какова вероятность того, что ее обратная грань - синяя?

**Задача 5. (Парадокс Монти-Холла)** 

Вы стали участником игры, в которой вам нужно выбрать одну из трёх дверей. За одной из дверей находится автомобиль, за двумя другими дверями — козы. Вы выбираете одну из дверей, например, номер 1, после этого ведущий, который знает, где находится автомобиль, а где — козы, открывает одну из оставшихся дверей, например, номер 3, за которой находится коза. После этого он спрашивает вас — не желаете ли вы изменить свой выбор и выбрать дверь номер 2? Увеличатся ли ваши шансы выиграть автомобиль, если вы примете предложение ведущего и измените свой выбор?

## ДЗ

**Задача 1. (1 балл)**

Партия продукции состоит из десяти изделий, среди которых два изделия дефектные. Какова вероятность того, что из пяти отобранных
наугад и проверенных изделий:
* ровно одно изделие дефектное;
* ровно два изделия дефектные;
* хотя бы одно изделие дефектное?

**Задача 2 (1 балл). (Парадокс дней рождений)** 

Найти вероятность того, что в классе из 23 человек, не менее двух учеников родились в один день.

**Задача 3 (1 балл).** 

На полке в случайном порядке расставлено
40 книг, среди которых находится трехтомник А. С. 
Пушкина. Найти вероятность того, что эти тома стоят в 
порядке возрастания слева направо (но не обязательно рядом).

**Задача 4 (1 балл). (Amazon)** 

Есть правильный кубик и неправильный кубик, грани которого мы не видим. Известно, что выпадение любой возможной суммы на этих кубиках равновероятно. Какие числа написаны на гранях неправильного кубика.


**Задача 5 (1 балл). (Boy or Girl paradox)** 

В семье есть 2 ребенка. Пол каждого ребенка - равно вероятен. Какова вероятность того, что в семье есть девочка, если
* В семье точно есть один мальчик
* В семье есть мальчик, который родился во Вторник.
